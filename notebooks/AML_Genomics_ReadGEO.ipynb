```python
import GEOparse
import pandas as pd
```

```python
# Define constants based on the paper
# DOI: https://doi.org/10.1016/j.isci.2019.100780

# SuperSeries GSE
geo_super_series = 'GSE122517'

# SubSeries GSE
geo_dataset1 = 'GSE122505'
geo_dataset2 = 'GSE122511'
geo_dataset3 = 'GSE122515'

# Save paths
dataset1_path = "./dataset1"
dataset2_path = "./dataset2"
dataset3_path = "./dataset3"
```

```python
gse_dataset1 = GEOparse.get_GEO(geo=geo_dataset1, destdir=dataset1_path)
gse_dataset2 = GEOparse.get_GEO(geo=geo_dataset2, destdir=dataset2_path)
gse_dataset3 = GEOparse.get_GEO(geo=geo_dataset3, destdir=dataset3_path)
```

```python
# The method for downloading suplementary files from GSE does not work atm.
# It generates the directory, but it is empty.
# Related issue: https://github.com/guma44/GEOparse/issues/72
# You can try it by yourself:
# gse_dataset1.download_supplementary_files()


# I use pandas to read the files from the link in the GSE file instead:

# The first link in the simplementary files, corresponds to the data matrix.
matrices_indx = 0

# We are working with tab separated values
separator = "\t"

# Read data and store in df (About 10 mins in a MacPro with bad WiFi...)
# We could use another library to load them. Maybe it is faster.
GSE122505_Dataset_1_matrix_df = pd.read_csv(gse_dataset1.metadata['supplementary_file'][matrices_indx], sep=separator)
GSE122511_Dataset_2_matrix_df = pd.read_csv(gse_dataset2.metadata['supplementary_file'][matrices_indx], sep=separator)
GSE122515_Dataset_3_ensembl_df = pd.read_csv(gse_dataset3.metadata['supplementary_file'][matrices_indx], sep=separator)
```
